```python
import glob

path = r'states' # use your path
all_files = glob.glob(path + "/*.csv")

df2=[]

for filename in all_files:
        temp_df=pd.read_csv((filename),encoding = "cp1252")
        temp_df=temp_df.loc[::,['S.No.','Item','2014-15']]
        temp_df['State']=filename[7:]
        df2.append(temp_df)
       
data2 = pd.concat(df2)
data2['State'] = data2['State'].map(lambda x: x.rstrip('.csv'))
data2['State'] = data2['State'].map(lambda x: x.title())
data2['Sector'] = np.nan

data2.loc[(data2.index[0:7]), 'Sector'] = "Primary"
data2.loc[(data2.index[7:11]), 'Sector'] = "Secondary"
data2.loc[(data2.index[11:27]), 'Sector'] = "Tertiary"
data2.tail(10)

dataB1 = data2
```
